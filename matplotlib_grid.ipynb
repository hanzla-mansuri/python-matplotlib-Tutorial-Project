```python
import matplotlib.pyplot as plt
import numpy as np
```

# adding grid line to plot

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 325])

plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Average oxygen", fontdict=font2)
plt.ylabel("Our calories", fontdict =font2)
plt.title("Health Monitor",fontdict=font1)
plt.grid()
plt.show()
```

# Now we will specify which grid lines to display via x axis or y axis

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Average oxygen", fontdict=font2)
plt.ylabel("Our calories", fontdict =font2)
plt.title("Health Monitor",fontdict=font1)
plt.plot(x, y)
plt.grid(axis='y') # you can mention 'x' or 'y'
plt.show()
```

# Seeting up the line propeties of the grid

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Average oxygen", fontdict=font2)
plt.ylabel("Our calories", fontdict =font2)
plt.title("Health Monitor",fontdict=font1)
plt.plot(x, y, ls = 'dashed', marker = 's', mfc = 'k', ms ='10')
plt.grid(axis='y', color = 'green', ls = 'dotted') 
plt.show()
```
