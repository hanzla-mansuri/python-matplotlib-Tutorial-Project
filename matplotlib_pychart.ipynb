#Creating The Pie chart . Function pie()

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#simple pie chart
y = np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
# add labels
y = np.array([35,25,25,15])
mylabels = ["apple", "banana", "orange","cherry"]
plt.pie(y, labels=mylabels)
plt.show()
```

```python
# change start angle , as defult is at 3:00 clock moving to anticlockwise
y = np.array([35,25,25,15])
mylabels = ["apple", "banana", "orange","cherry"]
plt.pie(y, labels=mylabels, startangle=90)
plt.show()
```

```python
# explode value
y = np.array([35,25,25,15])
mylabels = ["apple", "banana", "orange","cherry"]
myexplode = [0,0.2,0,0]
plt.pie(y, labels=mylabels,explode=myexplode)
plt.show()
```

```python
# The shadow parameter for creativity with color of wedges
y = np.array([35,25,25,15])
mylabels = ["apple", "banana", "orange","cherry"]
myexplode = [0.2,0,0,0]
mycolors = ['black', 'hotpink','blue','green']
plt.pie(y, labels=mylabels,explode=myexplode,shadow=True, colors= mycolors)
plt.legend()
plt.show()
```

```python
#
y = np.array([35,25,25,15])
mylabels = ["apple", "banana", "orange","cherry"]
myexplode = [0.2,0,0,0]
mycolors = ['black', 'hotpink','blue','green']
plt.pie(y, labels=mylabels,explode=myexplode,shadow=True, colors= mycolors)
plt.legend(title = "Fruites")
plt.show()
```
