# Display mutliple multiple subplot () you can draw multiple plots in one figure

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plot 1 & Plot 2

```python
x  =np.array([0, 1 , 2, 3])
y  =np.array([3, 8, 1, 10])
plt.subplot(1, 2, 2) # The figure has 1 row , 2 columns and this is 1 (first) plot
plt.plot(x,y, color ='k')

x  =np.array([1, 2, 2, 3])
y  =np.array([10, 20, 30, 40])
plt.subplot(1, 2, 1) # The figure has 1 row , 2 columns and this is 2 (2nd) plot
plt.plot(x,y, color = 'r')

plt.show()
```

# Plot 1 and Plot top of each other

```python
x  =np.array([0, 1 , 2, 3])
y  =np.array([3, 8, 1, 10])
plt.subplot(2, 1, 1) # The figure has 2 row , 1 columns and this is 1 (first) plot
plt.plot(x,y, color ='k')

x  =np.array([1, 1, 2, 3])
y  =np.array([10, 20, 30, 40])
plt.subplot(2, 1, 2) # The figure has 1 row , 1 columns and this is 2 (2nd) plot
plt.plot(x,y, color = 'r')

plt.show()
```

# now we will draw challage of 6 plot

```python
# Plot 1
x  =np.array([0, 1 , 2, 3])
y  =np.array([3, 8, 1, 10])
plt.subplot(2, 3, 1) # The figure has 3 row , 3 columns and this is 1 plot
plt.plot(x,y, color ='k')
plt.title('Plot 1')
# Plot 2
x  =np.array([2, 3, 2, 3])
y  =np.array([10, 20, 60, 40])
plt.subplot(2, 3, 2) # The figure has 3 row , 3 columns and this is 2 plot
plt.plot(x,y, color = 'r')
plt.title('Plot 2')
# Plot 3
x  =np.array([2, 3, 1, 3])
y  =np.array([10, 20, 25, 40])
plt.subplot(2, 3, 3) # The figure has 3 row , 3 columns and this is 3 plot
plt.plot(x,y, color = 'k')
plt.title('Plot 3')
# Plot 4
x  =np.array([2, 3, 2, 6])
y  =np.array([10, 20, 30, 40])
plt.subplot(2, 3, 4) # The figure has 3 row , 3 columns and this is 4 plot
plt.plot(x,y, color = 'b')
plt.title('Plot 4')
# Plot 5
x  =np.array([2, 3, 2, 3])
y  =np.array([10, 20, 30, 45])
plt.subplot(2, 3, 5) # The figure has 3 row , 3 columns and this is 5 plot
plt.plot(x,y, color = 'g')
plt.title('Plot 5')
# Plot 6
x  =np.array([2, 3, 2, 3])
y  =np.array([10, 5, 30, 45])
plt.subplot(2, 3, 6) # The figure has 3 row , 3 columns and this is 6 plot
plt.plot(x,y, color = 'hotpink')
plt.title('Plot 6')
plt.suptitle("Haana Plots")
plt.show()
```
