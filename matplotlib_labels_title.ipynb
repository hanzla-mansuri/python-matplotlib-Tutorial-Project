```python
# Creat labels for a plot
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 325])

plt.plot(x,y)
plt.xlabel("Average oxygen", color = "g")
plt.ylabel("Our calories", color = "b")
plt.title("Health Monitor", color = "r", fontsize = '20')
plt.show()
```

```python
# font style with varable via fontdic
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 325])

plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Average oxygen", fontdict=font2)
plt.ylabel("Our calories", fontdict =font2)
plt.title("Health Monitor",fontdict=font1)
plt.show()
```

```python
# use LOC to change locaiton of title and labes
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 325])

plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Average oxygen", fontdict=font2)
plt.ylabel("Our calories", fontdict =font2)
plt.title("Health Monitor",fontdict=font1, loc = 'left')
plt.show()
```
