```python
country = []
new_cases = []
dates = []
with open('WHO-COVID-19-global-data.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    
    for row in csv_reader:
        if row['Country'] == 'India':
            # Check if 'New_cases' is present and convertible to an integer
            if 'New_cases' in row and row['New_cases'].isdigit():
                country.append(row['Country'])
                new_cases.append(int(row['New_cases']))
                dates.append(row['Date_reported'])

# Print the extracted data for India
# for i in range(len(country)):
#     print(f"Date: {dates[i]}, Country: {country[i]}, New Cases: {new_cases[i]}")
plt.plot(dates,new_cases)
```
