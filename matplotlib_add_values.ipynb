MatplotLib Tutorial || How to Add Data Values in the Bar or Line Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sample data
catagoreis = ['cat1', 'cat2', 'cat3']
values = [20,35,45]
```

```python
#bar Plot

plt.bar(catagoreis, values)

# adding labels by for loop

for i, value in enumerate(values):
    plt.text(i,value-5,str(value),ha='center',va='bottom')
    print(i, value)
```

```python
#line chart
x_values = np.arange(1,11,1)
y_values = [10,25,13,32,20,30,25,22,15,19]

plt.plot(x_values,y_values, marker='o')
for x, y in zip(x_values,y_values):
    plt.text(x+0.2,y,str(y),ha = 'left', va = 'top')
```
