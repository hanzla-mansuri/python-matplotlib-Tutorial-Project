# Matplotlib OOP style

```python
import matplotlib.pyplot as plt
import numpy as np  
```

```python
x = np.linspace(10,100,100)
y = np.random.binomial(10,0.5,100)
fig, ((ax,ax2) )= plt.subplots(2,1)


# plot

ax.hist(y, bins=range(12), align='left', rwidth=0.5)
ax.set_title('ax')
ax2.plot(x,y, marker = 'o', color = "b", mfc = 'r')
ax2.set_title('ax2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)  # 100 evenly spaced points from 0 to 10

print(f"Lenght is {len(x)} for x")
# Define the functions to plot
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x)
y4 = np.log(x + 1)

# Create a figure with 4 subplots
fig, ((ax1, ax3), (ax2, ax4)) = plt.subplots(2, 2, figsize=(10, 8))

# Plot each function on a separate subplot
ax1.plot(x, y1, color='r',marker = 's', ms=5, mec = 'k', mfc ='hotpink')
ax1.set_title('Sine Function')

ax2.plot(x, y2, color='g')
ax2.set_title('Cosine Function')

ax3.plot(x, y3, color='b')
ax3.set_title('Exponential Function')

ax4.plot(x, y4, color='m')
ax4.set_title('Logarithmic Function')

# Adjust layout
plt.tight_layout()

# Super title
plt.suptitle("My learning plots", c = "orange")

# Display the plot
plt.show()
```
